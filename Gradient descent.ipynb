```python
curr_val=2
learning_rate=0.01
precision=0.000001
prev_step_val=1
max_iteration=1000
iteration=0
df=lambda x:(x+3)**2
```

```python
def f(x):
     return(x+3)**2
```

```python
def df(x):
    return 2*(x+3)
```

```python
values=[]
```

```python
while precision<prev_step_val and iteration<max_iteration:
    prev_val=curr_val
    curr_val=curr_val-learning_rate*df(prev_val)
    prev_step_val=abs(curr_val-prev_val)
    iteration+=1
    print("Iteration :",iteration,"value :",curr_val)
    values.append(curr_val)
```

```python
print("Local minima at :",curr_val)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(values)
plt.xlabel("Iteration")
plt.ylabel("values")
plt.title("Gradient descent Algorithm")
plt.show()
```
